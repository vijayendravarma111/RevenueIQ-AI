```python
import pandas as pd
import numpy as np

superstore = pd.read_csv(
    "../data/processed/superstore_clean.csv",
    encoding="latin1"
)

products = (
    superstore[
        ["Product ID", "Product Name"]
    ]
    .drop_duplicates()
)

np.random.seed(42)

inventory = pd.DataFrame()

inventory["product_id"] = products["Product ID"]

inventory["current_stock"] = np.random.randint(
    50,
    1000,
    len(products)
)

inventory["reorder_level"] = np.random.randint(
    20,
    200,
    len(products)
)

inventory["supplier_lead_time"] = np.random.randint(
    2,
    30,
    len(products)
)

inventory.to_csv(
    "../data/external/inventory_data.csv",
    index=False
)

inventory.head()
```

```python
import pandas as pd
import numpy as np

superstore = pd.read_csv(
    "../data/processed/superstore_clean.csv",
    encoding="latin1"
)

products = (
    superstore[
        ["Product ID", "Product Name", "Sales"]
    ]
    .groupby(
        ["Product ID", "Product Name"]
    )
    .mean()
    .reset_index()
)

np.random.seed(42)

competitors = pd.DataFrame()

competitors["product_id"] = products["Product ID"]

competitors["competitor_name"] = np.random.choice(
    [
        "Amazon",
        "Walmart",
        "Target",
        "BestBuy"
    ],
    len(products)
)

competitors["our_price"] = (
    products["Sales"]
    .round(2)
)

competitors["competitor_price"] = (
    competitors["our_price"] *
    np.random.uniform(
        0.9,
        1.1,
        len(products)
    )
).round(2)

competitors["capture_date"] = pd.Timestamp.today().date()

competitors.to_csv(
    "../data/external/competitor_data.csv",
    index=False
)

competitors.head()
```
